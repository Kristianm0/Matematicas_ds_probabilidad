# Dependencias

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-5,5, num=N)
```

## Función lineal
$$y=mx+b$$

```python
# creamos la funcion
def f(x):
  return x

#lo mostramos en pantalla
plt.plot(x, f(x))
plt.grid()
```

## Función escalón o de Heaviside
$$H(x) =
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
#funcion
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y

#variables
N=1000

#mostramos en pantalla
y = H(x)

plt.plot(x,y)
plt.grid()
```

## Función sigmoide

$$f(x)=\frac{1}{1-e^{-x}}$$

```python
#funcion
def f(x):
  return 1/(1 + np.exp(-x))

#variables
N=1000

#mostramos en pantalla
y = f(x)

plt.plot(x,y)
plt.grid()
```

## Función tangente hiperbólica

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

```python
#funcion
def f(x):
  return np.tanh(x)

#variable
N=1000

#mostramos en pantalla
y = f(x)

plt.plot(x,y)
plt.grid()
```

## Función ReLU

$$R(x)=max(0,x)$$

```python
#funcion
def f(x):
  return np.maximum(x,0)

# variables
N=1000

#mostramos en pantalla
y = f(x)

plt.plot(x,y)
plt.grid()
```
